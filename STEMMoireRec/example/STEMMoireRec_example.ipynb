```python
smh = SMR.userinput.read_dm3('Silicon.dm3')
```

```python
SMR.view.imview(smh.data)
```

```python
# Information about the crystal structure
SMR.userinput.crystal_info(smh, 543.1, 543.1, 543.1, 90, 90, 90, 'FCC')
# Experimental resolution and sampling base
SMR.userinput.exp_info(smh, 120, [[1, 1, 0], [0, 0, 1], [1, -1, 0]])
```

```python
list_resolved_allowed_reflections = SMR.crystal.list_reflections(smh, 10)
print('Allowed and resolved reflections')
for reflection in list_resolved_allowed_reflections:
    print(reflection['hkl'])
print('Allowed and resolved reflections in sampling base')
for reflection in smh.crystal_reflections:
    print(reflection['hkl'])
```

```python
moire_reflections = SMR.moiresim.simulate_moire_sampling(smh)
SMR.view.imview_sim(smh)
```

```python
masks_param = SMR.moirerec.set_masks_param(50, 25, smh)
SMR.view.imview_mask(masks_param, smh.data)
```

```python
moire_masked = SMR.moirerec.masking_moire_ft(masks_param, smh)
SMR.view.imview(np.log10(np.abs(moire_masked)), vmin=0)
```
